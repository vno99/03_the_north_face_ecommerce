## Import librairies

```python
import sys

import pandas as pd
```

```python
data = pd.read_csv("../data/data_cleaned.csv")
```

```python
class ClusterError(Exception):
    pass

def find_similar_items(item_id):
    try:
        article = data[data["id"] == int(item_id)]
        selected_cluster = article["clusters"].iloc[0]

        if selected_cluster == -1:
            raise ClusterError

        return (selected_cluster,
                sorted(data[ (data["clusters"] == selected_cluster) & 
                            (data["id"] != int(item_id)) ].sample(n=5)["id"].to_list()))
    
    except (KeyError, IndexError) as e:
        print(f"Article {item_id} non trouvé")
        
    except ValueError:
        print("Veuillez saisir un numéro d'article")

    except ClusterError:
        print("L'article sélectionné n'est dans aucun cluster")
    
```

```python
item_id = input("Donne moi un item_id. \n Je te suggèrerais d'autres articles similaires.")

try:
    result = find_similar_items(item_id)

    if result is None:
        sys.exit()
    
    cluster, suggestion_list = result
    selected_article = data[data["id"] == int(item_id)]
    selected_article_desc = selected_article["description"].loc[0]
    print(f"Produit sélectionné : {item_id} - {selected_article_desc[:50]} ...")
    print(f"Cluster : {cluster} \n")


    suggested_articles = data[data["id"].isin(suggestion_list)]
    print("Suggestion de produits: ")

    for suggested_article_id in suggestion_list:
        suggested_article = data[data["id"] == suggested_article_id]
        suggested_article_desc = suggested_article["description"].iloc[0]
        print(f"{suggested_article_id:>3} - {suggested_article_desc[:50]} ...")

except SystemExit:
    pass
```
